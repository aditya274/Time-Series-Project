## Removing outliers using Inter Quartile Range (IQR)

  * To remove outliers, we first find the IQR of each commodity for same date.
            IQR = Q3 - Q1
  * We find lower and upper whiskers (range) using the formula :
            lower whisker = Q1 - IQR*1.5 
            upper whisker = Q3 + IQR*1.5 
  * We remove any value that is either lower than lower whisker or higher than upper whisker.
  

```python
def IQR(df):
    Q1 = df["modal_price"].quantile(0.25)
    Q3 = df["modal_price"].quantile(0.75)
    IQR = Q3 - Q1
    lower_whis = Q1 - IQR*1.5 # Lower whisker
    upper_whis = Q3 + IQR*1.5 # Upper whisker
    to_delete = ( (df["modal_price"] < lower_whis) | (df["modal_price"] > upper_whis) )  
    # Returning "True" for outliers.
    return to_delete
```

```python
outlier_indices = []
for commodity in titled_commodities:
    comm = data[data['Commodity'] == commodity]
    comm_dates = comm['date'].unique()
    for date in comm_dates:
        df = comm.loc[comm['date'] == date]
        bool_list = IQR(df)
        to_delete = bool_list.loc[bool_list == True].index.to_list() # Extracting outliers index.
        outlier_indices.extend(to_delete)  # Keeping record of outliers index.
```

```python
clean_data = data.drop(outlier_indices)  # Removing outliers from datasets.
clean_data.to_csv('dataset/clean_data.csv', index=False)  # And saving it.
```
